Drop Useless Columns

```python
# Removing columns that are not useful for our ML model
# These columns mostly contain text, URLs, or timestamps.

remove_cols = [
    'name','screen_name','description','location',
    'url','profile_image_url','profile_banner_url',
    'profile_image_url_https','profile_background_image_url',
    'profile_background_image_url_https',
    'created_at','updated','dataset','utc_offset', 'profile_background_tile'
]


for col in remove_cols:
    if col in df.columns:
        df.drop(col, axis=1, inplace=True)

print("Remaining columns after removing unnecessary ones:\n")
print(df.columns.tolist())
```

Fix Missing Values

```python
#Handle missing values

for col in df.columns:
    if df[col].dtype in ['int64','float64']:
        df[col] = df[col].fillna(df[col].median())
    elif df[col].dtype == 'bool':
        df[col] = df[col].fillna(False).astype(int)
    else:
        df[col] = df[col].fillna("missing")

print("Remaining NaNs:", df.isnull().sum().sum())
```

```python
# Fix boolean columns properly

if 'verified' in df.columns:
    df['verified'] = df['verified'].fillna(0)
    df['verified'] = df['verified'].astype(int)

if 'protected' in df.columns:
    df['protected'] = df['protected'].fillna(0)
    df['protected'] = df['protected'].astype(int)

print("Remaining NaNs:", df.isnull().sum().sum())
```

Outlier Handling

```python
#Handling outliers using Winsorization (1% - 99%)

# Selecting numeric columns except the target column
numeric_cols = [col for col in df.columns
                if df[col].dtype in ['int64', 'float64'] and col != 'is_fake']

print("Numeric columns being clipped:", numeric_cols)

for col in numeric_cols:
    # Calculate lower (1%) and upper (99%) limits
    lower_limit = df[col].quantile(0.01)
    upper_limit = df[col].quantile(0.99)

    # Clip values outside the limits
    df[col] = df[col].clip(lower=lower_limit, upper=upper_limit)

print("\nOutlier handling (Winsorization) completed.")
```

Feature Engineering

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class FeatureEngineer(BaseEstimator, TransformerMixin):
    def __init__(self):
        pass

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        X = X.copy()

        # engineered features
        X['followers_per_friend'] = X['followers_count'] / (X['friends_count'] + 1)
        X['favourites_per_status'] = X['favourites_count'] / (X['statuses_count'] + 1)
        X['friends_per_status'] = X['friends_count'] / (X['statuses_count'] + 1)

        X['activity_score'] = (
            X['followers_count'] +
            X['friends_count'] +
            X['statuses_count']
        )

        X['popularity_score'] = (
            X['followers_count'] * 0.6 +
            X['friends_count'] * 0.3 +
            X['listed_count'] * 0.1
        )

        # Handling inf/NaN
        X.replace([np.inf, -np.inf], 0, inplace=True)
        X.fillna(0, inplace=True)

        return X
```

Define X and y

```python
# Defining target and features
if 'id' in df.columns:
    df.drop('id', axis=1, inplace=True)

y = df['is_fake']
X = df.drop('is_fake', axis=1)

print("Features (X) Shape:", X.shape)
print("Target (y) Shape:", y.shape)
```

Train/Test Split

```python
#Split data into train and test sets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=42, stratify=y
)

print("Train shape:", X_train.shape)
print("Test shape:", X_test.shape)
print("\nClass distribution in training set:")
print(y_train.value_counts())
```

```python
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()
```

Build Preprocessing Pipeline

```python
# Preprocessing Pipeline

from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer

# Identify feature types
numeric_cols = X_train.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = X_train.select_dtypes(include=['object']).columns.tolist()

print("Numeric features:", numeric_cols)
print("Categorical features:", categorical_cols)

# Numeric pipeline
numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Categorical pipeline (will run only if we have categorical columns)
categorical_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

# Combined preprocessing steps
preprocessor = ColumnTransformer([
    ('num', numeric_pipeline, numeric_cols),
    ('cat', categorical_pipeline, categorical_cols)
])
```

Build Model Pipelines

```python
# Integrate Preprocessing with ML Algorithms

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.pipeline import Pipeline

# Logistic Regression
pipe_log = Pipeline([
    ('feature_engineering', FeatureEngineer()),
    ('preprocess', preprocessor),
    ('model', LogisticRegression(max_iter=2000))
])

# Random Forest
pipe_rf = Pipeline([
    ('feature_engineering', FeatureEngineer()),
    ('preprocess', preprocessor),
    ('model', RandomForestClassifier(n_estimators=100, max_depth=5, random_state=42))
])

# Decision Tree
pipe_dt = Pipeline([
    ('feature_engineering', FeatureEngineer()),
    ('preprocess', preprocessor),
    ('model', DecisionTreeClassifier(max_depth=4, random_state=42))
])

# SVM
pipe_svm = Pipeline([
    ('feature_engineering', FeatureEngineer()),
    ('preprocess', preprocessor),
    ('model', SVC(kernel='linear', probability=True, random_state=42))
])
```

```python
# Define the models dictionary
models = {
    "Logistic Regression": pipe_log,
    "Random Forest": pipe_rf,
    "Decision Tree": pipe_dt,
    "SVM": pipe_svm
}
```

```python
from sklearn.model_selection import cross_val_score

print("\n Cross Validation Results (5-Fold) \n")

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=5, scoring='accuracy')
    print(f"{name}: {scores.mean():.4f} ± {scores.std():.4f}")
```

Train ALL 4 models

```python
# Train all 4 models

pipe_log.fit(X_train, y_train)
pipe_rf.fit(X_train, y_train)
pipe_dt.fit(X_train, y_train)
pipe_svm.fit(X_train, y_train)

print("Training complete for Logistic Regression, Random Forest, Decision Tree, and SVM.")
```

TESTING & MAKING PREDICTIONS (Using X_test)

```python
#Testing and making predictions on X_test

predictions = {}    # store predicted labels
probabilities = {}  # store predicted probabilities (for ROC-AUC)

for name, model in {
    "Logistic Regression": pipe_log,
    "Random Forest": pipe_rf,
    "Decision Tree": pipe_dt,
    "SVM": pipe_svm
}.items():

    # Predictions
    y_test_pred = model.predict(X_test)

    # Probabilities
    y_test_prob = model.predict_proba(X_test)[:, 1]  # probability of class 1 (fake)

    # Save
    predictions[name] = y_test_pred
    probabilities[name] = y_test_prob
```

```python
import pandas as pd
import joblib

# Load best saved model (with feature engineering inside pipeline)
model = joblib.load("models/random_forest.joblib")

print("\n")
print("  FAKE ACCOUNT DETECTION TEST UI")
print("\n")

# --- RAW USER INPUTS (no feature engineering needed now) ---
followers = int(input("Enter Followers Count: "))
friends = int(input("Enter Friends Count: "))
statuses = int(input("Enter Statuses Count: "))
favourites = int(input("Enter Favourites Count: "))
listed = int(input("Enter Listed Count: "))

verified = int(input("Verified? (0 = No, 1 = Yes): "))
protected = int(input("Protected Account? (0 = No, 1 = Yes): "))
geo_enabled = int(input("Geo Enabled? (0 = No, 1 = Yes): "))
default_profile = int(input("Default Profile? (0 = No, 1 = Yes): "))
default_profile_image = int(input("Default Profile Image? (0 = No, 1 = Yes): "))
profile_use_background_image = int(input("Using Background Image? (0 = No, 1 = Yes): "))

# Build input DataFrame — ONLY RAW FEATURES
data = {
    "followers_count": followers,
    "friends_count": friends,
    "statuses_count": statuses,
    "favourites_count": favourites,
    "listed_count": listed,

    "verified": verified,
    "protected": protected,
    "geo_enabled": geo_enabled,
    "default_profile": default_profile,
    "default_profile_image": default_profile_image,
    "profile_use_background_image": profile_use_background_image,
}

df_input = pd.DataFrame([data])

# --- MODEL PREDICTION ---
prediction = model.predict(df_input)[0]

# --- DISPLAY RESULT ---
print("\n")
if prediction == 1:
    print(" This account is predicted as: FAKE")
else:
    print(" This account is predicted as: REAL")
print("\n")
```
